```python
print("Rozpakowywanie CelebA (chwilę to potrwa)...")
!unzip -q celeba-dataset.zip -d ./celeba-dataset
print("Rozpakowywanie zakończone.")
```

Kopiowanie danych do GCS

```python
print(f"Rozpoczynam kopiowanie CelebA do gs://{BUCKET_CELEBA_ORIGINAL} ...")
!gsutil -m -q cp -r ./celeba-dataset gs://{BUCKET_CELEBA_ORIGINAL}
print("Wszystko skopiowane do GCS!")
```
