```python
print(f"Rozpoczynam kopiowanie LFW do gs://{BUCKET_LFITW_ORIGINAL}/lfw_dataset ...")
!gsutil -m cp -r ./lfw_dataset gs://{BUCKET_LFITW_ORIGINAL}
print("Wszystko skopiowane do GCS!")
```
